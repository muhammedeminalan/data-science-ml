```python
import numpy as np
from scipy import stats

control = np.array([6.8, 7.1, 6.5, 7.0, 6.9, 6.6, 7.2, 6.7, 6.8, 6.9])
experiment = np.array([7.4, 7.6, 7.1, 7.5, 7.3, 7.7, 7.2, 7.6, 7.4, 7.5])

t_stat, p_two_sided = stats.ttest_ind(experiment, control, equal_var=False)

# Tek yönlü test (deney > kontrol) için p-değerini yarıya indiriyoruz
p_one_sided = p_two_sided / 2 if t_stat > 0 else 1.0

alpha = 0.05
print("t:", t_stat, "p(one-sided):", p_one_sided)

if p_one_sided < alpha:
    print("H0 reddedilir: Yeni tarif memnuniyeti artırıyor gibi görünüyor.")
else:
    print("H0 reddedilemez: Artış için yeterli kanıt yok.")
```

```python
import math

success_a, n_a = 120, 1000
success_b, n_b = 150, 980

p_a = success_a / n_a
p_b = success_b / n_b

# pooled proportion
p_pool = (success_a + success_b) / (n_a + n_b)

se = math.sqrt(p_pool * (1 - p_pool) * (1/n_a + 1/n_b))
z = (p_b - p_a) / se

# Tek yönlü p-değeri (B > A)
# Normal dağılım CDF için yaklaşık hesap
def norm_cdf(x):
    return (1.0 + math.erf(x / math.sqrt(2.0))) / 2.0

p_value = 1 - norm_cdf(z)

print("pA:", p_a, "pB:", p_b, "z:", z, "p(one-sided):", p_value)
```
